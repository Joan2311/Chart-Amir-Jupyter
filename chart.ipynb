```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

productA = np.array([120, 130, 125])
productB = np.array([150, 160, 170])
productC = np.array([100, 110, 105])
months = np.array(["Jan", "Feb", "Mar"])

sale = {"month":months,
       "product A":productA,
        "product B":productB,
        "product C": productC
       }
g = pd.DataFrame(sales)
print(g)
print(g.describe)
```

# mean():
